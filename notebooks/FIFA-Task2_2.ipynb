## Understanding the Amount of null values in the DataFrame

```python
df.isnull().sum()
```

```python
def get_null_percentages(df, columns):
    result = {}
    for col in columns:
        if col in df.columns:
            null_count = df[col].isnull().sum()
            total_count = len(df)
            null_percentage = (null_count / total_count) * 100
            result[col] = round(null_percentage, 2)
        else:
            result[col] = "Column not found"

    return result

columns_checking=['Goal_assist', 'Main_Time','Playing_Position','Competition']
null_percentage = get_null_percentages(df, columns_checking)

for column, percentage in null_percentage.items():
    if isinstance(percentage, float):
        print(f"{column}: {percentage}%")
    else:
        print(f"{column}: {percentage}")

```

```python
sns.barplot(data=null_percentage, x=null_percentage.values(), y=null_percentage.keys(), palette='viridis', hue=null_percentage.keys()).set(title="Percentage of Missing Values", xlabel="Percentage (%)", ylabel="Features")
plt.show()
```

```python
def fill_goal_assist_with_mode(df, competition_col, assist_col):
    
    mode_values = df.groupby(competition_col)[assist_col].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
    
    def fill_na(row):
        if pd.isna(row[assist_col]):
            return mode_values[row[competition_col]]
        else:
            return row[assist_col]

    df[assist_col] = df.apply(fill_na, axis=1)
    return df

df_filled = fill_goal_assist_with_mode(df, 'Competition', 'Goal_assist')
        
        
```
